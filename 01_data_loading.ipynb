```python
# Setting base dirsctory for datasets

import os
import numpy as np
import pandas as pd
import seaborn as sns

BASE_DIR = os.getenv('CBIS_DDSM_PATH', r'C:\Users\aryan\OneDrive\Desktop\DATASET')
if not os.path.exists(BASE_DIR):
    raise FileNotFoundError(f"Dataset directory {BASE_DIR} does not exist")

# Set plotting style
sns.set_theme(style='whitegrid')
```

```python

#Path to csv file
csv_dir = os.path.join(BASE_DIR, 'csv')
if not os.path.exists(csv_dir):
    raise FileNotFoundError(f"CSV file {csv_dir} does not exist")

calc_train_path = os.path.join(csv_dir, 'calc_case_description_train_set.csv')
mass_train_path = os.path.join(csv_dir, 'mass_case_description_train_set.csv')
calc_test_path = os.path.join(csv_dir, 'calc_case_description_test_set.csv')
mass_test_path = os.path.join(csv_dir, 'mass_case_description_test_set.csv')
dicom_info_path = os.path.join(csv_dir, 'dicom_info.csv')
meta_path = os.path.join(csv_dir, 'meta.csv')

# Load the CSV file into a DataFrame

try:
    calc_train_df = pd.read_csv(calc_train_path)
    mass_train_df = pd.read_csv(mass_train_path)
    calc_test_df = pd.read_csv(calc_test_path)
    mass_test_df = pd.read_csv(mass_test_path)
    dicom_info_df = pd.read_csv(dicom_info_path)
    meta_df = pd.read_csv(meta_path)
except FileNotFoundError as e:
    raise FileNotFoundError(f"Error loading CSV files: {e}")

# View of calcification training dataset
print("Calcalification Training Dataset:")
display(calc_train_df.head())

# View of mass training dataset
print("Mass Training Dataset:")
display(mass_train_df.head())
```

```python
# Checking the distribution pathology in mass and calc datasets

print('Calcification Pathology Distribution(train):')
display(calc_train_df['pathology'].value_counts())

print('Mass Pathology Distribution(train):')
display(mass_train_df['pathology'].value_counts())

# Plotting distribution

import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.countplot(data = calc_train_df, x = 'pathology', ax = axes[0])
axes[0].set_title('Calcification Pathology Distribution (Train):')
axes[0].set_xlabel('Pathology')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis = 'x', rotation= 45)

sns.countplot(data = mass_train_df, x = 'pathology', ax = axes[1])
axes[1].set_title('Mass Pathology Distribution (Train):')
axes[1].set_xlabel('Pathology')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis = 'x', rotation = 45)
plt.tight_layout()
plt.show()
```
